```python
import pandas as pd
```

```python
dataset = pd.read_csv('weight-height.csv')
```

```python
dataset
```

```python
X = dataset['Height']
y = dataset['Weight']
```

```python
import keras
from keras.models import Sequential
```

Like Linear Regression is a model (algo), same in deep learning, Sequential is also an model 

```python
model = Sequential()
```

```python
from keras.layers import Dense
```

Dense works for the Neural layers, to add more layers
activation => which kind of algorithm you want to use (linear classification or binary classification)

```python
model.add(Dense( activation="linear", input_shape=(1, ), units=1, kernel_initializer="zeros" ,
                bias_initializer="zeros" ))
#Input shape => how many features you have in your X variable
```

Bias Initializer = Kernel Initializer

```python
model.summary()
```

```python
model.get_config()
```

```python
model.get_weights()
```

```python
from keras.optimizers import Adam
model.compile(loss='mean_squared_error', optimizer=Adam())
```

```python
model.fit(X,y, epochs=1000)
```

```python
model.get_weights()
```

```python
model.get_weights()
```

```python
model.predict([73])
```

```python
model.save("my_weight_model.h5")
#created and saved the model
```

```python
#to import and directly use the model

from keras.models import load_model
m = load_model('my_weight_model.h5')
m.predict([73])
```
